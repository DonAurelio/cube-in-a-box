```python
%load_ext autoreload
from satsearch import Search
import json
from ls_public_bucket import make_stac_metadata_doc
```

```python
lon_min = -180
lon_max = 180
lat_min = -90
lat_max = 90

start_date = '2020-02-01'
end_date = '2020-03-01'

extent = [lon_min, lat_min, lon_max, lat_max]

srch = Search().search(
        bbox=extent,
        datetime='{}T00:00:00Z/{}T23:59:59Z'.format(start_date, end_date),
        collections=["sentinel-s2-l2a-cogs"],
    )
```

```python
srch.found()
```

```python
for item in srch.items():
    stac_item = item
    print(item)
    break
```
